```python
import pandas as pd

# Crear el dataset
data = {
    'Producto': ['Camiseta', 'Pantalón', 'Zapatos', 'Camiseta', 'Zapatos', 'Pantalón', 'Camiseta'],
    'Cantidad': [10, 5, 8, 7, 3, 6, 2],
    'Precio': [15.0, 40.0, 60.0, 15.0, 60.0, 40.0, 15.0],
    'Fecha': pd.to_datetime([
        '2024-07-01', '2024-07-01', '2024-07-02',
        '2024-07-03', '2024-07-03', '2024-07-04', '2024-07-05'
    ])
}

df = pd.DataFrame(data)

# 1. Producto más vendido (sumando cantidades)
ventas_por_producto = df.groupby('Producto')['Cantidad'].sum()
producto_mas_vendido = ventas_por_producto.idxmax()
cantidad_mas_vendida = ventas_por_producto.max()

print(f"Producto más vendido: {producto_mas_vendido} con {cantidad_mas_vendida} unidades.")

# 2. Productos ordenados por precio de mayor a menor (precio promedio por producto)
precio_por_producto = df.groupby('Producto')['Precio'].mean().sort_values(ascending=False)

print("\nProductos ordenados por precio de mayor a menor:")
print(precio_por_producto)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear el dataset
data = {
    'Producto': ['Camiseta', 'Pantalón', 'Zapatos', 'Camiseta', 'Zapatos', 'Pantalón', 'Camiseta'],
    'Cantidad': [10, 5, 8, 7, 3, 6, 2],
    'Precio': [15.0, 40.0, 60.0, 15.0, 60.0, 40.0, 15.0],
    'Fecha': pd.to_datetime([
        '2024-07-01', '2024-07-01', '2024-07-02',
        '2024-07-03', '2024-07-03', '2024-07-04', '2024-07-05'
    ])
}

df = pd.DataFrame(data)

# Agrupar por producto y sumar cantidades
ventas_por_producto = df.groupby('Producto')['Cantidad'].sum().sort_values(ascending=False)

# Crear gráfico de columnas
plt.figure(figsize=(8,5))
ventas_por_producto.plot(kind='bar', color='purple')
plt.title('Cantidad total vendida por producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
df = pd.DataFrame(data)

# KPI 1: Total de unidades vendidas
total_unidades = df['Cantidad'].sum()

# KPI 2: Ingreso total
df['Ingreso'] = df['Cantidad'] * df['Precio']
ingreso_total = df['Ingreso'].sum()

# KPI 3: Producto más vendido
ventas_por_producto = df.groupby('Producto')['Cantidad'].sum()
producto_mas_vendido = ventas_por_producto.idxmax()
cantidad_mas_vendida = ventas_por_producto.max()

# KPI 4: Ingreso promedio por venta (transacción)
ingreso_promedio = ingreso_total / len(df)

# KPI 5: Precio promedio por producto
precio_promedio_producto = df.groupby('Producto')['Precio'].mean()

# Mostrar KPIs
print(f"Total de unidades vendidas: {total_unidades}")
print(f"Ingreso total: ${ingreso_total:.2f}")
print(f"Producto más vendido: {producto_mas_vendido} con {cantidad_mas_vendida} unidades")
print(f"Ingreso promedio por venta: ${ingreso_promedio:.2f}")
print("\nPrecio promedio por producto:")
print(precio_promedio_producto)
```
